```python
cc = ss.to_dict()
```

```python
for i in cc['features']:
    #print(cc['features'][i])
    #print(cc['features'][i]['properties'])
    cc['features'][i]['properties'].pop('OBJECTID')
    cc['features'][i]['properties'].pop('Changerooms')
    cc['features'][i]['properties'].pop('FacilityAge')
    cc['features'][i]['properties'].pop('FacilityCondition')
    cc['features'][i]['properties'].pop('FacilityUpgradeAge')
    cc['features'][i]['properties'].pop('Facility_ID')
    cc['features'][i]['properties'].pop('FaciltySportPlayedID')
    cc['features'][i]['properties'].pop('FieldSurfaceType')
    cc['features'][i]['properties'].pop('LGA')
    cc['features'][i]['properties'].pop('NumberFieldCourts')
    cc['features'][i]['properties'].pop('StreetName')
    cc['features'][i]['properties'].pop('StreetType')
    cc['features'][i]['properties'].pop('StreetNo')
    cc['features'][i]['properties'].pop('SuburbTown')
    cc['features'][i]['properties']['marker-size']= 'medium'
    cc['features'][i]['properties']['marker-color']= '#3bb2d0'
    cc['features'][i]['properties']['marker-symbol']= 'star'
```

```python
geojson = []
for i in cc['features']:
    geojson.append(cc['features'][i])
```
